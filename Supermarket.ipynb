# graphs

```python
sns.histplot(df['Rating'], kde=False, color='red', linestyle='--')
plt.axvline(x=df['Rating'].mean(), color='blue', linestyle='--')
plt.axvline(x=np.percentile(df['Rating'],25),color = 'g', linestyle = '--')
plt.axvline(x = np.percentile(df['Rating'],75),color = 'y',linestyle = '--')
plt.show()
```
